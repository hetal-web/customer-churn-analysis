```python
import pandas as pd 
import numpy as np 
```

```python
np.random.seed(42)
n= 400
```

```python
df  = pd.DataFrame({
    "Age" : np.random.randint(18,60,n),
    "Monthly_Charges" : np.random.randint(200,2000,n),
    "Tenure" : np.random.randint(1,60,n),
    "Contract": np.random.choice(["Monthly","Yearly"],n),
    "Churn" : np.random.choice([0,1],n)
})

df.head()
```

```python
df["Churn"].mean()
```

```python
df.groupby("Contract")["Churn"].mean()
```

```python
df.groupby("Tenure")["Churn"].mean().head()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.countplot(x="Churn",data = df)
plt.title("Churn Distribution")
plt.show()
```

```python
sns.barplot(x="Contract", y= "Churn",data=df)
plt.title("Churn by Contract type")
plt.show()
```

```python
def churn_risk(monthly_charges,tenure,contract):
    if contract == "Monthly" and tenure <12:
        return "High Risk"
    elif monthly_charges > 1500:
        return "Medium Risk"
    else:
        return "Low Risk"   
print(churn_risk(1800,5,"Monthly")
)
```
